### Code 3

```python
base_url = "http://stat.snu.ac.kr/heeseok/teaching/asm17/"
html = urllib.request.urlopen(base_url)
html_contents = str(html.read())
```

```python
html_contents
```

```python
url_list = re.findall(r"ch[0-9].pdf", html_contents)
print(url_list)
```

```python
for url in url_list:
    file_name = url
    full_url = base_url + file_name
    print(full_url)
    fname, header = urllib.request.urlretrieve(full_url, file_name)
    print("End Download")
```
